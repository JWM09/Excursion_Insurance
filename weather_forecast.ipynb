```python
# Import Required Libraries
import pandas as pd
import requests
import json
import os
from dotenv import load_dotenv
from glom import glom
```

```python
# Load .env environment variables into the notebook
load_dotenv()
```

```python
hnl_api_key = os.getenv('HNL_Weather')
```

```python
type(hnl_api_key)
```

```python
#define th base request URL
hnl_url = 'https://api.openweathermap.org/data/3.0/onecall?lat=21.31&lon=-157.86&exclude=alerts,minutely,hourly,currents&appid='
```

```python
# concatenate with api key
hnl_weather_request = hnl_url + hnl_api_key
```

```python
# execute api call
weather_forecast = requests.get(hnl_weather_request).json()
```

```python
# Display Response
print(json.dumps(weather_forecast, indent=4, sort_keys=True))
```

```python
df = pd.DataFrame.from_dict(pd.json_normalize(weather_forecast), orient='columns')
df
```

```python
df.to_csv('weather_test', encoding='utf-8', index=False)
```
